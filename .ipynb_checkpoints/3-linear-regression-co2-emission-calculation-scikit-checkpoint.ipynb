## Select useful features from our dataset

```python
e = "ENGINESIZE"
c = "CYLINDERS"
f = "FUELCONSUMPTION_COMB"
co2 = "CO2EMISSIONS"
data = data[[e,c,f,co2]]
data.head()
```

## Plotting the Data

Plot the data on scatter plot to find out which feature can be used to make the predictions. Plotting each feature against CO2 Emissions.

### EngineSize vs CO2Emissions

```python
plt.scatter(data[e],data[co2])
plt.xlabel("No. of Cylinders")
plt.ylabel("Emissions")
plt.show()
```

### FuelConsumption vs CO2Emissions

```python
plt.scatter(data[f],data[co2])
plt.xlabel("No. of Cylinders")
plt.ylabel("Emissions")
plt.show()
```

Here we can see that we can easily plot a regression line in ENGINE SIZE VS CO2 EMISSION plot.

Now we will divide our dataset into 2 parts. One for training data and another for testing data. Well use 80% of the data for training and 20% of data to test our predictions.

```python
# list[start:stop]

# Here, :int(len(data) * 0.8) means "from the beginning up to (but not including) the index int(len(data) * 0.8)."
train = data[:int((len(data)*0.8))]

# Here, int(len(data) * 0.8): means "starting from the index int(len(data) * 0.8) to the end of the list."
test = data[(int(len(data)*0.8)):]

print(f"Training Data: {train.shape}")
print(f"Testing Data: {test.shape}")
print(f"Total Data: {data.shape}")
```

## Using SciKit Learn

```python
# modeling
# using sklearn project to model data

from sklearn import linear_model
import numpy as np
regression = linear_model.LinearRegression()

train_x = np.array(train[[e]])
train_y = np.array(train[[co2]])

regression.fit(train_x, train_y)

print(f"Co-Efficient (m): {regression.coef_[0][0]}") #slope
print(f"Intercept (c): {regression.intercept_[0]}") #intercept
```

### Plotting the Regression Line

```python
plt.scatter(train[e],train[co2],color='blue')
plt.plot(train_x, regression.coef_*train_x+regression.intercept_,'-r')
plt.xlabel("Engine Size")
plt.ylabel("Emission")
```

```python
plt.scatter(train[e],train[co2],color='blue')
plt.plot(train_x, regression.coef_*train_x+regression.intercept_,'-r')
plt.xlabel("Engine Size")
plt.ylabel("Emission")
```

### Predicting CO2 Emission

```python
def getRegressionPrediction(feature, slope, intercept):
    prediction = feature * slope + intercept
    return prediction
```

```python
engineSize = 3.5

estimatedEmission = getRegressionPrediction(engineSize, regression.coef_[0][0], regression.intercept_[0])
print(f"Estimated Emission: {estimatedEmission}")
```

```python
estimatedEmission = getRegressionPrediction(train[e], regression.coef_[0][0], regression.intercept_[0])
print(f"Estimated Emission: {estimatedEmission}")
print(f"Actual Emission: {train[co2]}")
```

## Performance Analysis

```python
from sklearn.metrics import r2_score

test_x = np.array(test[[e]])
test_y = np.array(test[[co2]])
test_y_=regression.predict(test_x)

print(f"Mean Absolute Error: {np.mean(np.absolute(test_y_ - test_y))}")
print(f"Mean Sum of Squares (MSE): {np.mean((test_y_-test_y)**2)}")
print(f"R2-Score: {r2_score(test_y_,test_y)}")
```
